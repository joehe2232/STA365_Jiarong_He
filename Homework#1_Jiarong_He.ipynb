# Q1

What proportion of students who answer this question quickly will pass the class?
Assume the probability of passing the class is 0.9.
Assume the probability of answering this question quickly is 0.6 if you'll pass the class,
while the probability drops to 0.3 if you'll not pass the class. 

\begin{align*}
\overset{posterior}{p(\theta|x)} &={} \frac{\overset{likelihood}{p(x|\theta)}\times \overset{prior}{p(\theta)}}{\overset{\text{evidence}}{p(x)}}
\end{align*}

We are given:

$$
P(\text{Pass}) = 0.9, 
\quad
P(\text{Not Pass}) = 1 - 0.9 = 0.1,
$$

$$
P(\text{Quick} \mid \text{Pass}) = 0.6,
\quad
P(\text{Quick} \mid \text{Not Pass}) = 0.3.
$$

We want to find:
$$
P(\text{Pass} \mid \text{Quick}).
$$

By Bayes' Theorem:
$$
P(\text{Pass} \mid \text{Quick}) 
= \frac{P(\text{Quick} \mid \text{Pass}) \; P(\text{Pass})}{P(\text{Quick})}.
$$

First, compute $$P(\text{Quick})$$ using the Law of Total Probability:
$$
P(\text{Quick}) 
= P(\text{Quick} \mid \text{Pass}) \, P(\text{Pass})
+ P(\text{Quick} \mid \text{Not Pass}) \, P(\text{Not Pass}).
$$

Substitute the values:
$$
P(\text{Quick})
= (0.6)(0.9) + (0.3)(0.1)
= 0.54 + 0.03
= 0.57.
$$

Then,
$$
P(\text{Pass} \mid \text{Quick}) 
= \frac{(0.6)(0.9)}{0.57} 
= \frac{0.54}{0.57} 
\approx 0.9474.
$$

Hence, approximately **\(94.74\%\)** of the students who answer quickly will pass the class.

# Q2

\begin{align*}
\text{Likelihood} && p(x|\theta,n) &={} \texttt{Multinomial}\\\\
\text{Prior} && p(\theta|\boldsymbol{\alpha}) &={} \texttt{Dirichlet} \\ 
\text{Posterior} && p(\theta|x,n,\boldsymbol{\alpha}) &={} \overset{\text{Bayes Theorem}}{\frac{p(x|\theta)p(\theta)}{p(x)}} = \; \texttt{What?}\\
&&& \quad\quad\quad\quad\quad\; = \; \texttt{What distribution?}\\
&&& \quad\quad\quad\quad\quad\quad\;\;\, \texttt{With what parameterization?}\\
\end{align*}

### _The posterior is always proportional to the joint_

$$\large p(\theta|x) = (1/c)\times p(\theta,x) \propto{} p(\theta,x)$$

**_and the posterior is always a proper distribution if the prior is_**

### Given:
- **Likelihood**: \\( p(\mathbf{x} \mid \boldsymbol{\theta}, n) = \text{Multinomial}(\mathbf{x}; n, \boldsymbol{\theta}) \\).
- **Prior**: \\( p(\boldsymbol{\theta} \mid \boldsymbol{\alpha}) = \text{Dirichlet}(\boldsymbol{\alpha}) \\), 
  where \\(\boldsymbol{\alpha} = (\alpha_1, \ldots, \alpha_K) \\).

We observe \\(\mathbf{x} = (x_1, \ldots, x_K)\\) counts of \\(K\\) categories (with \\(\sum_{i=1}^K x_i = n\\)) and wish to determine the posterior distribution:

\\[
p(\boldsymbol{\theta} \mid \mathbf{x}, n, \boldsymbol{\alpha})
= \frac{p(\mathbf{x} \mid \boldsymbol{\theta}, n) \, p(\boldsymbol{\theta} \mid \boldsymbol{\alpha})}{p(\mathbf{x})}.
\\]

### Deriving the Posterior

1. **Likelihood** (Multinomial):  
   \\[
   p(\mathbf{x} \mid \boldsymbol{\theta}, n) 
   = \frac{n!}{\prod_{i=1}^K x_i !} 
     \prod_{i=1}^K \theta_i^{x_i}.
   \\]

2. **Prior** (Dirichlet):  
   \\[
   p(\boldsymbol{\theta} \mid \boldsymbol{\alpha})
   = \frac{1}{B(\boldsymbol{\alpha})}
     \prod_{i=1}^K \theta_i^{\alpha_i - 1},
   \\]
   where 
   \\[
     B(\boldsymbol{\alpha}) 
     = \frac{\prod_{i=1}^K \Gamma(\alpha_i)}{\Gamma\!\bigl(\sum_{i=1}^K \alpha_i\bigr)}
   \\]
   is the multinomial beta function (normalizing constant).

3. **Unnormalized Posterior**:  
   \\[
   p(\boldsymbol{\theta} \mid \mathbf{x}, n, \boldsymbol{\alpha})
   \; \propto \;
   p(\mathbf{x} \mid \boldsymbol{\theta}, n) 
   \; p(\boldsymbol{\theta} \mid \boldsymbol{\alpha}).
   \\]
   Substituting the forms,
   \\[
   \propto
   \left(
       \frac{n!}{\prod_{i=1}^K x_i !}
       \prod_{i=1}^K \theta_i^{x_i}
   \right)
   \left(
       \frac{1}{B(\boldsymbol{\alpha})}
       \prod_{i=1}^K \theta_i^{\alpha_i - 1}
   \right).
   \\]
   Ignoring constants that do not depend on \\(\boldsymbol{\theta}\\), we have
   \\[
   \propto
   \prod_{i=1}^K \theta_i^{x_i + \alpha_i - 1}.
   \\]

4. **Recognizing the Dirichlet Form**:  
   The posterior is of the form
   \\[
   \prod_{i=1}^K \theta_i^{(\alpha_i + x_i) - 1},
   \\]
   which is exactly a Dirichlet distribution with parameters 
   \\[
   \bigl(\alpha_1 + x_1, \alpha_2 + x_2, \ldots, \alpha_K + x_K \bigr).
   \\]

### Conclusion

Hence the posterior distribution is:

\\[
\boxed{
p(\boldsymbol{\theta} \mid \mathbf{x}, n, \boldsymbol{\alpha})
= \text{Dirichlet}(\boldsymbol{\alpha} + \mathbf{x})
= \text{Dirichlet}
\bigl(\alpha_1 + x_1, \ldots, \alpha_K + x_K \bigr).
}
\\]

**Answer**: The posterior is a **Dirichlet** distribution, parameterized by \\(\boldsymbol{\alpha} + \mathbf{x}\\).

**_And the posterior is always a proper distribution if the prior is a Dirichlet distribution._**


# Q3

Create a sample from a three category multinomial distribution and visualize the multinomial distribution and the multinomial sample; then, visualize the simplex Dirichlet distribution of the prior and the posterior using a ternary plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import mpltern

# 1. Define a three-category multinomial distribution
p = [0.2, 0.5, 0.3]  # Probabilities for each category
n = 20              # Number of trials

# 2. Draw one random sample from the multinomial
x = np.random.multinomial(n, p)  # e.g., something like [4, 11, 5]
print("Observed sample from Multinomial(n=20, p=[0.2, 0.5, 0.3]):", x)

# 3. Visualize the theoretical distribution vs. the observed counts
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Left: Theoretical distribution (probabilities)
ax[0].bar(range(len(p)), p, color=["C0", "C1", "C2"])
ax[0].set_xticks(range(3))
ax[0].set_xticklabels([f"Category {i+1}" for i in range(3)])
ax[0].set_ylabel("Probability")
ax[0].set_title("Theoretical Multinomial Probabilities")

# Right: Observed counts from the single sample
ax[1].bar(range(len(x)), x, color=["C0", "C1", "C2"])
ax[1].set_xticks(range(3))
ax[1].set_xticklabels([f"Category {i+1}" for i in range(3)])
ax[1].set_ylabel("Count")
ax[1].set_title("Observed Multinomial Sample")
plt.tight_layout()
plt.show()

# 4. Define a Dirichlet prior, then form the posterior.
#    Let's choose a simple prior alpha = [1, 1, 1] (a uniform prior on the 3-simplex).
alpha_prior = np.array([1.0, 1.0, 1.0])      # Prior parameters
alpha_posterior = alpha_prior + x           # Posterior parameters

print("Prior alpha:", alpha_prior)
print("Posterior alpha:", alpha_posterior)

# 5. Sample from both the prior and the posterior Dirichlet distributions for visualization.
num_samples = 1000
samples_prior = np.random.dirichlet(alpha_prior, size=num_samples)       # [N, 3]
samples_posterior = np.random.dirichlet(alpha_posterior, size=num_samples)

# 6. Visualize the prior and posterior on a ternary plot.
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='ternary')  # Ternary projection from mpltern

# Plot prior samples
ax.scatter(
    samples_prior[:, 0], samples_prior[:, 1], samples_prior[:, 2],
    color="blue", alpha=0.3, s=10, label="Prior Samples"
)

# Plot posterior samples
ax.scatter(
    samples_posterior[:, 0], samples_posterior[:, 1], samples_posterior[:, 2],
    color="red", alpha=0.3, s=10, label="Posterior Samples"
)

# Some additional labels and decorations
ax.set_title("Dirichlet Prior vs Posterior Samples (Ternary Plot)")
ax.legend(loc="upper right", bbox_to_anchor=(1.3, 1.05))
ax.grid(True)

# Ternary axes labels
ax.set_tlabel("Category 1")
ax.set_llabel("Category 2")
ax.set_rlabel("Category 3")

plt.show()
```
