### Q2: Continue "Q2" of the previous weeks 3 and 4

1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

2. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a non normal distribution and your prior for $\tau$ is a non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

3. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a yet another different again non normal distribution and your prior for $\tau$ is a yet another different again non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

---

```python
# Case 1: Normal theta and Gamma tau

# -----------------------------
# 1. Simulate some data
# -----------------------------
# For reproducibility
np.random.seed(42)

# True parameters
true_theta = 2.0
true_tau = 1.0  # precision, so sigma = 1/sqrt(tau) = 1
theta0 = 0

# Simulate data: n = 100 observations
n = 100
draws = 10000
data = stats.norm(loc=true_theta, scale=1/np.sqrt(true_tau)).rvs(size=n)

# -----------------------------
# 2. Define the PyMC model
# -----------------------------
with pm.Model() as model:
    
    theta_prior_n = 1  # strength of prior belief as units of data
    tau0 = theta_prior_n*true_tau
    tau_SS_prior = 1  # prior belief regarding true_sum_of_squares
    tau_prior_n = 1  # strength of prior belief as units of data
    
    # Prior for theta
    theta = pm.Normal("theta", mu=theta0, sigma=tau0**(-0.5))
    
    # Prior for tau
    tau = pm.Gamma("tau", alpha=tau_prior_n/2, beta=tau_SS_prior/2)
    
    # Likelihood: data ~ Normal(theta, sigma)
    y = pm.Normal("y", mu=theta, tau=tau, observed=data)
    
    # -----------------------------
    # 3. Sample from the posterior
    # -----------------------------
    trace = pm.sample(draws=10000, chains=2, tune=100, return_inferencedata=True, target_accept=0.95)

# -----------------------------
# 4. Diagnostic Assessments
# -----------------------------
az.plot_trace(trace, combined=True)
plt.tight_layout()

print(az.summary(trace, round_to=5))

az.plot_forest(trace, var_names=["theta", "tau"], 
               combined=False, hdi_prob=0.95, r_hat=True);
```

```python
# Case 2: Cauchy theta and Lognormal tau

# -----------------------------
# 1. Simulate some data
# -----------------------------
# For reproducibility
np.random.seed(42)

# True parameters
true_theta = 2.0
true_tau = 1.0  # precision, so sigma = 1/sqrt(tau) = 1
theta0 = 0

# Simulate data: n = 100 observations
n = 100
draws = 10000
data = stats.norm(loc=true_theta, scale=1/np.sqrt(true_tau)).rvs(size=n)

# -----------------------------
# 2. Define the PyMC model
# -----------------------------
with pm.Model() as model:
    
    # Non-normal prior for theta: use a Cauchy distribution
    theta = pm.Cauchy("theta", alpha=0, beta=5)
    
    # Non-gamma prior for tau: use a LogNormal distribution
    tau = pm.Lognormal("tau", mu=0, sigma=1)
    
    # Convert precision tau to standard deviation sigma for the Normal likelihood.
    sigma = 1 / pm.math.sqrt(tau)
    
    # Likelihood: data ~ Normal(theta, sigma)
    y = pm.Normal("y", mu=theta, sigma=sigma, observed=data)
    
    # -----------------------------
    # 3. Sample from the posterior
    # -----------------------------
    trace = pm.sample(draws=10000, chains=2, tune=100, return_inferencedata=True, target_accept=0.95)

# -----------------------------
# 4. Diagnostic Assessments
# -----------------------------
az.plot_trace(trace, combined=True)
plt.tight_layout()

print(az.summary(trace, round_to=5))

az.plot_forest(trace, var_names=["theta", "tau"], 
               combined=False, hdi_prob=0.95, r_hat=True);
```

```python
# Case 3: Student-t distributed theta and Halfnormal tau

# -----------------------------
# 1. Simulate some data
# -----------------------------
# For reproducibility
np.random.seed(42)

# True parameters
true_theta = 2.0
true_tau = 1.0  # precision, so sigma = 1/sqrt(tau) = 1
theta0 = 0

# Simulate data: n = 100 observations
n = 100
draws = 10000
data = stats.norm(loc=true_theta, scale=1/np.sqrt(true_tau)).rvs(size=n)

# -----------------------------
# 2. Define the PyMC Model
# -----------------------------
with pm.Model() as model:
    
    # Non-normal prior for theta: Student's t distribution with 3 degrees of freedom
    theta = pm.StudentT("theta", nu=3, mu=0, sigma=10)
    
    # Non-gamma prior for tau: Half-Normal distribution ensuring positivity
    tau = pm.HalfNormal("tau", sigma=10)
    
    # Convert precision to standard deviation for the likelihood
    sigma = 1 / pm.math.sqrt(tau)
    
    # Likelihood: observed data modeled as normally distributed
    y = pm.Normal("y", mu=theta, sigma=sigma, observed=data)
    
    # -----------------------------
    # 3. Sample from the posterior
    # -----------------------------
    trace = pm.sample(draws=10000, chains=2, tune=100, return_inferencedata=True, target_accept=0.95)

# -----------------------------
# 4. Diagnostic Assessments
# -----------------------------
az.plot_trace(trace, combined=True)
plt.tight_layout()

print(az.summary(trace, round_to=5))

az.plot_forest(trace, var_names=["theta", "tau"], 
               combined=False, hdi_prob=0.95, r_hat=True);
```

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    
    # find interval of grid points where f(x_grid) > y 
    # then extend the enterval so endpoints f(a)<y and f(b)<y 
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid
    
    x_ = a + stats.uniform().rvs()*(b-a)    
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f" 
    elif x_ < x:  # or if "x_ was above f on the left side of the interval" 
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):
    
    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)
    
    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

# Slice Sampling and Its Use in a Gibbs Sampler

In many Bayesian applications we need to sample from a target density
$$
f(x)
$$
that is known only up to a normalizing constant. **Slice sampling** is an MCMC method that does this by sampling uniformly from the region under the curve of \\( f(x) \\). Below we explain (1) how the given Markov algorithm for slice sampling works and (2) how slice sampling can replace a Metropolis–Hastings (MH) step in a Metropolis-within-Gibbs sampler when the full conditionals are only known up to a constant.

---

## Part 1: How the Markov Algorithm for Slice Sampling Works

Consider the univariate target density \\( f(x) \\). The slice sampling algorithm proceeds as follows:

1. **Vertical (Auxiliary) Step:**  
   Given the current state \\( x^{(t)} \\), sample an auxiliary variable
   $$
   y \sim \operatorname{Uniform}\Big(0, f\big(x^{(t)}\big)\Big).
   $$
   This \\( y \\) defines a horizontal level (or “slice”) under the density curve.

2. **Horizontal (Slice) Step:**  
   Define the slice as the set
   $$
   S = \{ x : f(x) > y \}.
   $$
   That is, \\( S \\) contains all \\( x \\)-values for which the density exceeds \\( y \\); in other words, it is the region *under the curve* \\( f(x) \\) above the level \\( y \\).

3. **Finding a Candidate by Interval Sampling and Shrinking:**  
   In practice, we may not know \\( S \\) exactly. The provided algorithm does the following:
   
   - **Locate an Interval:** Using a grid of \\( x \\)-values (e.g., `x_grid`), it finds an interval \\([a, b]\\) where \\( f(x) > y \\). The endpoints are then slightly extended so that \\( f(a) < y \\) and \\( f(b) < y \\).
   
   - **Draw a Candidate:** A candidate \\( x' \\) is drawn uniformly from \\([a, b]\\).
   
   - **Accept or Shrink:**  
     - If \\( f(x') > y \\), then \\( x' \\) lies in the slice and is accepted immediately.
     - If not, the interval is **shrunk**: if \\( x' < x^{(t)} \\), then the left endpoint is moved to \\( x' \\); if \\( x' > x^{(t)} \\), the right endpoint is moved to \\( x' \\). The algorithm then repeats (recursively) the candidate draw within the new, smaller interval until a valid \\( x' \\) is found.

This procedure yields a Markov chain whose stationary distribution is proportional to \\( f(x) \\) even if the normalization constant is unknown.

---

## Part 2: Using Slice Sampling in a Gibbs Sampler

In a Gibbs sampler, we often need to sample from a full conditional distribution,
$$
p(\theta|\cdot) \propto f(\theta),
$$
where \\( f(\theta) \\) is known only up to a constant. A typical Metropolis–Hastings update would require proposing a candidate and then accepting or rejecting it based on an acceptance probability that involves the unknown constant. Instead, slice sampling offers an alternative update mechanism:

1. **Initial Value:**  
   Begin with the current state \\( \theta^{(t)} \\) of the parameter. This is the *initial value* for the slice sampling update.

2. **Vertical Step:**  
   Sample an auxiliary variable:
   $$
   y \sim \operatorname{Uniform}\Big(0, f\big(\theta^{(t)}\big)\Big).
   $$
   Here, the **curve** we are sampling beneath is \\( f(\theta) \\)—the unnormalized density of the full conditional.

3. **Define the Slice:**  
   The slice is defined as:
   $$
   S = \{ \theta : f(\theta) > y \}.
   $$
   This represents all values of \( \theta \) under the curve \( f(\theta) \) that lie above the level \( y \).

4. **Horizontal Step:**  
   Draw a candidate \\( \theta^{(t+1)} \\) uniformly from \\( S \\). As in Part 1, if \\( S \\) is not explicitly known, we use an interval method (finding an interval and then shrinking it) to sample from \\( S \\).

5. **Automatic Acceptance:**  
   Because the candidate is sampled uniformly from the region where \\( f(\theta) > y \\), it is automatically accepted. There is no need for an additional MH acceptance step.

Thus, slice sampling replaces the MH step by providing an efficient way to update the parameter from its full conditional even when the density is known only up to a constant.
